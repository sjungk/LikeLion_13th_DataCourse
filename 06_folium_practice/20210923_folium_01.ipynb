```python
# 우리의 핫 플레이스 (강사님 코드)
loc = folium.Map(location=[37.544613, 127.039464], 
                 zoom_start=12)

folium.Marker([37.539840, 127.070459],
               popup='건대역5번출구',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc)  # 건대역 5번 출구
folium.Marker([37.512706500791, 127.10257850098388],
               popup='롯데 월드 타워',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc) # 롯데 월드 타워
folium.Marker([37.572392, 126.959524],
              popup='독립문',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc)    # 독립문
folium.Marker([33.1164523, 126.2672144],
               popup='마라도',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc)  # 마라도
folium.Marker([37.55603, 126.97235],
               popup='서울역',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc)  # 서울역
folium.Marker([37.51383570, 127.104168],
               popup='롯데 월드 몰',
              icon = folium.Icon(color='purple', icon='info-sign')
              ).add_to(loc)  # 롯데 월드 몰

loc.save("13th_hotplace.html")

loc
```
