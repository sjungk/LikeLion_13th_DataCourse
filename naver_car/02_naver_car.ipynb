```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import numpy as np
driver = webdriver.Chrome('chromedriver')
```

```python
# 전체 시판 모델을 데이터화 시켜보자
```

```python
# 현대 캐스퍼 : //*[@id="content"]/div[3]/ul/li[1]/div
# 기아 EV6 : //*[@id="content"]/div[3]/ul/li[3]/div

# i30 : //*[@id="content"]/div[3]/ul/li[8]/div
```

```python
basic_url = "https://auto.naver.com/car/mainList.nhn?isMnfcoAll=Y&page="
url3 = "https://auto.naver.com/car/mainList.nhn"

cars = []
for i in range(1,8,1):
    real_url = basic_url + str(i)
    page = urlopen(real_url)
    soup = BeautifulSoup(page, 'lxml')
    sell_car_all = soup.find_all("span", class_="box")
    
   
    for one in sell_car_all:
        cars.append(one.text)
    print(cars)
```

```python
dict_dat = {"현재 시판 중인 자동차" : cars}
dat = pd.DataFrame(dict_dat)
dat.to_csv("네이버_현재_시판차(국내).csv", index=True)
dat.to_excel("네이버_현재_시판차(국내).xlsx", index=True)
```

```python
# 단종 모델을 알아보자
basic_url2 = "https://auto.naver.com/car/mainList.nhn?mnfcoNo=0&modelType=DC&order=0&importYn=N&page="
url4 = "https://auto.naver.com/car/mainList.nhn?mnfcoNo=0&modelType=DC&order=0&importYn=N"

no_cars=[]
for i in range(1,60,1):
    real_url2 = basic_url2 + str(i)
    page = urlopen(real_url2)
    soup = BeautifulSoup(page, 'lxml')
    no_sell_car = soup.find_all("span", class_="box")
    
    
    for one in no_sell_car:
        no_cars.append(one.text)
    print(no_cars)    
```

```python
dict_dat = {"단종된 자동차" : no_cars}
dat = pd.DataFrame(dict_dat)
dat.to_csv("네이버_단종_자동차(국내).csv", index=True)
dat.to_excel("네이버_단종_자동차(국내).xlsx", index=True)
```
