```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import numpy as np
driver = webdriver.Chrome('chromedriver')
```

```python
# 전체 시판 모델을 데이터화 시켜보자 (해외 편)
```

```python
# 2022 돈커부트 D8 GTO : //*[@id="content"]/div[3]/ul/li[1]/div
# 2022 쉐보레 실버라도 : //*[@id="content"]/div[3]/ul/li[2]/div
```

```python
basic_url = "https://auto.naver.com/car/mainList.nhn?importYn=Y&page="


cars = []
for i in range(1,31,1):
    real_url = basic_url + str(i)
    page = urlopen(real_url)
    soup = BeautifulSoup(page, 'lxml')
    sell_car_all = soup.find_all("span", class_="box")
    
   
    for one in sell_car_all:
        cars.append(one.text)
    print(cars)
```

```python
dict_dat = {"현재 시판 중인 자동차" : cars}
dat = pd.DataFrame(dict_dat)
dat.to_csv("네이버_현재_시판차(해외).csv", index=True)
dat.to_excel("네이버_현재_시판차(해외).xlsx", index=True)
```

```python
# 단종 모델을 알아보자
basic_url2 = "https://auto.naver.com/car/mainList.nhn?mnfcoNo=0&modelType=DC&order=0&importYn=Y&page="

no_cars=[]
for i in range(1,31,1):
    real_url2 = basic_url2 + str(i)
    page = urlopen(real_url2)
    soup = BeautifulSoup(page, 'lxml')
    no_sell_car = soup.find_all("span", class_="box")
    
    
    for one in no_sell_car:
        no_cars.append(one.text)
    print(no_cars)    
```

```python
dict_dat = {"단종된 자동차" : no_cars}
dat = pd.DataFrame(dict_dat)
dat.to_csv("네이버_단종_자동차(해외).csv", index=True)
dat.to_excel("네이버_단종_자동차(해외).xlsx", index=True)
```
